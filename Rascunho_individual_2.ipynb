```python
import pandas as pd
import altair as alt


url = 'https://raw.githubusercontent.com/altair-viz/vega_datasets/master/vega_datasets/_data/stocks.csv'
df = pd.read_csv(url)
df.head()
```

```python

line_with_tooltip = alt.Chart(df).mark_line().encode(
    x='date:T',
    y='price:Q',
    color='symbol:N',
    tooltip=['symbol:N', 'date:T', 'price:Q']
).properties(
    title='Preço das Ações com Tooltip'
)

line_with_tooltip.show()
```

```python

scatter_chart = alt.Chart(df).mark_point().encode(
    x='date:T',
    y='price:Q',
    size='price:Q',
    color='symbol:N',
    tooltip=['symbol:N', 'date:T', 'price:Q']
).properties(
    title='Gráfico de Dispersão com Tamanho e Cor'
)

scatter_chart.show()
```

```python

facet_chart = alt.Chart(df).mark_line().encode(
    x='date:T',
    y='price:Q',
    color='symbol:N'
).facet(
    column='symbol:N'
).properties(
    title='Preço das Ações por Empresa'
)

facet_chart.show()
```

```python
import pandas as pd
import altair as alt

url = "https://raw.githubusercontent.com/altair-viz/vega_datasets/master/vega_datasets/_data/stocks.csv"
df = pd.read_csv(url)

df['d'] = pd.to_datetime(df['date'])
df['a'] = df['d'].dt.year

pm = df.groupby(['a', 'symbol'])['price'].mean().reset_index()

graf = alt.Chart(pm).mark_bar().encode(
    x=alt.X('a:O', title='Ano'),
    y=alt.Y('price:Q', title='Preço Médio'),
    color=alt.Color('symbol:N', title='Empresa', legend=alt.Legend(title='Empresa')),
    tooltip=['a', 'symbol', 'price']
).properties(
    title='Preço Médio Anual das Ações por Empresa',
    width=600,
    height=400
).interactive()

graf.show()
```
